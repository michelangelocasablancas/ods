[Association Rules Generation from Frequent Itemsets](https://rasbt.github.io/mlxtend/user_guide/frequent_patterns/association_rules/)

```python
import pandas as pd
```

```python
tweets_ods = pd.read_feather('../data/tweets_ods')
tweets_ods.head()
```

```python
ods = tweets_ods.drop(['text', 'index'], axis=1)
ods.head()
```

El dataframe `ods` ya tiene el formato de _cesta de la compra_

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
frequent_itemsets = apriori(ods, min_support=0.04, use_colnames=True)

frequent_itemsets
```

```python
frequent_itemsets.sort_values(by='support', ascending=False).head(20)
```
