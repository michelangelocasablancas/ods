```python
tweets_ods = pd.read_feather('../data/tweets_ods')
tweets_ods
```

```python
usuario_ods = pd.DataFrame([])
```

```python
usuario_ods['usuario'] = tweets_es['nom_completo_usuario']
```

```python
usuario_ods['num_ods'] = tweets_ods.iloc[:, 2:].dropna().apply(sum, axis=1)
```

```python
usuario_ods.head()
```

```python
grouped = usuario_ods.groupby('usuario').sum()
grouped.head()
```

```python
grouped.sort_values('num_ods', ascending=False, inplace=True)
grouped
```
